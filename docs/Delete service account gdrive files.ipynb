```python
from google.oauth2 import service_account
import googleapiclient
import googleapiclient.discovery

scopes = ["https://www.googleapis.com/auth/drive"]
drive_api_service_name = "drive"
drive_api_version = "v3"
GCP_SERVICE_ACCOUNT_JSON = {}# YOUR_SERVICE_ACCOUNT_CONTENT_IN_JSON in dict
credentials = service_account.Credentials.from_service_account_info(
    GCP_SERVICE_ACCOUNT_JSON,
    scopes=scopes,
)

drive_client = googleapiclient.discovery.build(
    drive_api_service_name,
    drive_api_version,
    credentials=credentials,
    cache_discovery=False,
)

a = drive_client.files().list(pageSize=1000).execute()
files = a.get("files")

for f in files:
    try:
        drive_client.files().delete(fileId=f.get("id")).execute()
    except Exception as e:
        print(e)

a = drive_client.files().emptyTrash().execute()
```
